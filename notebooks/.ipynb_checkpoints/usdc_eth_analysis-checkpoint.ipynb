## ETH Price Data


```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv("../data/usdc_raw.csv")
df.head()
```

```python


# Convert value to actual USDC amount
df["usdc_amount"] = df["value"].astype(float) / 1_000_000

# Convert timestamp to date
df["date"] = pd.to_datetime(df["timeStamp"], unit="s").dt.date

# Deduplicate overlapping transactions
df = df.drop_duplicates(subset="hash")

# Aggregate to daily USDC volume
daily_usdc = (
    df.groupby("date")["usdc_amount"]
      .sum()
      .reset_index()
      .rename(columns={"usdc_amount": "usdc_volume"})
)

daily_usdc.head()
```

```python
# CoinGecko endpoint for ETH daily prices
url = "https://api.coingecko.com/api/v3/coins/ethereum/market_chart"
params = {
    "vs_currency": "usd",
    "days": 30,      # match your USDC window
    "interval": "daily"
}

response = requests.get(url, params=params)
data = response.json()

# Extract prices
prices = data["prices"]

eth_prices = pd.DataFrame(prices, columns=["timestamp", "eth_price"])
eth_prices["date"] = pd.to_datetime(eth_prices["timestamp"], unit="ms").dt.date
eth_prices = eth_prices[["date", "eth_price"]]

eth_prices.head()
```

```python
# Merge daily USDC volume with ETH prices
merged = pd.merge(
    daily_usdc,
    eth_prices,
    on="date",
    how="inner"
)

merged.head()
```

```python
merged.shape
```

```python
merged
```

```python


fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot USDC volume (left axis)
ax1.plot(merged["date"], merged["usdc_volume"], label="USDC Volume")
ax1.set_xlabel("Date")
ax1.set_ylabel("USDC Volume")
ax1.tick_params(axis="x", rotation=45)

# Plot ETH price (right axis)
ax2 = ax1.twinx()
ax2.plot(merged["date"], merged["eth_price"], color="orange", label="ETH Price")
ax2.set_ylabel("ETH Price (USD)")

plt.title("Daily USDC Transaction Volume and ETH Price")
plt.tight_layout()
plt.show()
```

Figure 1:Daily USDC Transaction Volume and ETH Price

This figure plots daily USDC transaction volume on Ethereum alongside the ETH/USD price. While both series exhibit substantial day-to-day variation, there is no visually strong contemporaneous correlation between ETH price levels and stablecoin transaction volume. This suggests that stablecoin usage may reflect liquidity and transactional demand rather than speculative price movements.



```python

# Define our regression variables
X = merged["eth_price"]
y = merged["usdc_volume"]   

# Add constant (intercept)
X = sm.add_constant(X)

# Fit regression
model = sm.OLS(y, X).fit()
```

```python
model.summary()
```

Interpretation of Regression Results

The OLS regression shows no statistically significant relationship between daily ETH prices and daily USDC transaction volume on Ethereum over the sample period. The estimated coefficient on ETH price is small and statistically insignificant (p = 0.907), and the model explains virtually none of the variation in USDC activity (RÂ² = 0.001).

This result suggests that stablecoin usage may not be directly driven by short-run movements in cryptocurrency prices. Instead, USDC transactions likely reflect payment activity, liquidity management, and institutional usage rather than speculative price dynamics.

```python
# Compute daily returns / growth rates
merged["eth_return"] = merged["eth_price"].pct_change()
merged["usdc_growth"] = merged["usdc_volume"].pct_change()

merged[["date", "eth_return", "usdc_growth"]].head()
```

```python
# Drop rows with NaNs from percent change
reg_data = merged.dropna(subset=["eth_return", "usdc_growth"])
reg_data.shape

```

```python
reg_data[["eth_return", "usdc_growth"]].describe()
```

```python
import statsmodels.api as sm

X = reg_data["eth_return"]
X = sm.add_constant(X)

y = reg_data["usdc_growth"]

model_returns = sm.OLS(y, X).fit()
model_returns.summary()
```

Returns-Based Regression Results

To address scale differences and potential non-stationarity, both ETH prices and USDC transaction volumes were transformed into daily percentage changes. The regression estimates whether short-run movements in ETH prices are associated with changes in stablecoin usage.

The results indicate no statistically significant relationship between ETH returns and USDC transaction growth over the sample period. This suggests that stablecoin usage is largely driven by transactional and liquidity-related factors rather than short-term speculative price movements in the broader cryptocurrency market.

The finding is consistent with the role of stablecoins as a medium of exchange and unit of account within crypto markets, rather than as speculative assets.
