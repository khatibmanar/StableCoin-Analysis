## Phase 2: Market Stress and Stablecoin Usage
## I. Motivation
This notebook extends the exploratory analysis in Phase 1 by examining whether stablecoin transaction activity responds systematically to periods of heightened cryptocurrency market volatility. While Phase 1 relied on raw Etherscan API data and was constrained by transaction caps and missing observations, Phase 2 uses indexed blockchain data from Dune Analytics to construct a more complete and consistent time series of USDC transaction activity on Ethereum.

The analysis focuses on aggregate on-chain measures of stablecoin usage, including transaction volume and transaction counts, and relates these outcomes to external indicators of cryptocurrency market stress, such as realized ETH price volatility. The objective is descriptive rather than causal: to assess whether stablecoins exhibit usage patterns consistent with a “flight-to-safety” or transactional role during volatile market conditions.

## II. Data and Measurement

## II.1 Stablecoin Transaction Data
Daily USDC transaction data are constructed using curated ERC-20 transfer records from Dune Analytics. Specifically, I query the tokens.transfers table, which provides fully decoded token-level transfer information, including transfer amounts and timestamps, without the truncation and missing-day issues present in raw API data.

The sample is restricted to:

    - Ethereum mainnet transfers (blockchain = 'ethereum')

    - USDC transactions (symbol = 'USDC')

    - The most recent three-year period

For each day, I construct two measures of stablecoin usage:

    - Daily USDC transfer volume, defined as the sum of USDC transferred across all transactions on that day

    - Daily USDC transaction count, defined as the number of USDC transfer events on that day

```python
# load Phase 2 USDC dataset
import pandas as pd

# Load Phase 2 USDC data exported from Dune
df_usdc = pd.read_csv("../data/usdc_daily_last_3_years.csv")

# Parse dates
df_usdc['day'] = pd.to_datetime(df_usdc['day'])

# Sort chronologically
df_usdc = df_usdc.sort_values('day')

# Basic checks
df_usdc.head(), df_usdc.tail(), df_usdc.describe()
```

The dataset contains one observation per day, spans the intended three-year period, and exhibits economically plausible magnitudes for transaction counts. However transaction volume seems to be read as a string

```python
# Force numeric conversion
df_usdc['daily_usdc_volume'] = pd.to_numeric(
    df_usdc['daily_usdc_volume'], errors='coerce'
)

# Check how many values became NaN
df_usdc['daily_usdc_volume'].isna().sum()
```

```python
# check if there are large outliers in the dataset
df_usdc['daily_usdc_volume'].describe()
```

```python
df_usdc.sort_values('daily_usdc_volume', ascending=False).head(10)
```

A small number of observations exhibit implausibly large transfer volumes ( exceeding 10^38), which are inconsistent with observed transaction counts and known stablecoin market sizes. These values are treated as corrupted data and excluded using a conservative upper bound on daily transfer volume.

```python
#No day in the sample should have USDC transfer volume exceeding $1 trillion
df_usdc = df_usdc[df_usdc['daily_usdc_volume'] < 1e12]
```

```python
# rerunning describe to check
df_usdc['daily_usdc_volume'].describe()
```

```python
#checking transaction counts remain unaffected
df_usdc['daily_usdc_tx_count'].describe()
```

The resulting dataset contains 919 daily observations and exhibits magnitudes consistent with known USDC market activity.

## II.2 Market Stress Measure 
To proxy for cryptocurrency market stress, I use ETH price volatility, computed from daily ETH prices obtained from external market data sources. Volatility is measured as realized volatility over a fixed window (e.g., rolling standard deviation of daily log returns), which captures short-term fluctuations in market conditions.

ETH volatility is merged with the on-chain USDC dataset by calendar date.


```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
# Download daily ETH-USD prices covering the same data range as our USDC data
eth = yf.download(
    "ETH-USD",
    start=df_usdc['day'].min(),
    end=df_usdc['day'].max(),
    interval="1d",
    auto_adjust=True
)

eth.head()
```

```python
# Keep closing price and reset index
eth = eth[['Close']].reset_index()
eth.rename(columns={'Date': 'day', 'Close': 'eth_price'}, inplace=True)

# Ensure date alignment
eth['day'] = pd.to_datetime(eth['day'])

eth.head()
```

```python
# Compute daily ETH log returns
eth['eth_log_return'] = np.log(eth['eth_price']).diff()

eth.head()
```

```python
# computing realized 7-day volatility 
eth['eth_vol_7d'] = eth['eth_log_return'].rolling(window=7).std()
```

```python
# Merge ETH volatility with USDC data 
df = pd.merge(
    df_usdc,
    eth[['day', 'eth_vol_7d']],
    on='day',
    how='inner'
)

df.head()
```

```python
# Flatten multi-level columns if they exist for the merge to work
eth.columns = eth.columns.get_level_values(0)
```

```python
eth.columns
```

```python
# drop the name column
eth.columns.name = None
```

```python
#verify no more multi-level columns
eth.columns
```

```python
# Merge ETH volatility with USDC data 
df = pd.merge(
    df_usdc,
    eth[['day', 'eth_vol_7d']],
    on='day',
    how='inner'
)

df.head(15)
```

```python
#drop intial NaN rows because volatility is constructed using a rolling window
df = df.dropna(subset=['eth_vol_7d'])
df['eth_vol_7d'].describe()
```

```python
import statsmodels.api as sm

# Dependent variable
y = df['daily_usdc_tx_count']

# Independent variable
X = df[['eth_vol_7d']]
X = sm.add_constant(X)

# OLS with heteroskedasticity-robust (HC1) standard errors
model = sm.OLS(y, X).fit(cov_type='HC1')

print(model.summary())
```

Baseline Regression Results

I estimate a univariate OLS regression of daily USDC transaction counts on same-day ETH volatility using heteroskedasticity-robust (HC1) standard errors. The estimated coefficient on ETH volatility is positive and statistically significant at the 1% level. Economically, a 0.01 increase in daily ETH volatility is associated with approximately 25,000 additional USDC transfer events, representing a meaningful increase relative to typical daily transaction counts.

The regression exhibits low explanatory power (R² = 0.063), which is expected given the parsimonious specification and the descriptive nature of the analysis. The regression residuals exhibit strong serial correlation, as indicated by a low Durbin–Watson statistic, which is a common feature in high-frequency transaction data. Furthermore, residual normality tests (Omnibus and Jarque–Bera) reject normality.


Overall, the results suggest that periods of heightened cryptocurrency market volatility are associated with increased stablecoin transaction activity, consistent with the view that stablecoins may serve as transactional or risk-mitigating instruments during periods of market stress.

```python
import numpy as np

# Creating log daily USDC volume
df['log_usdc_volume'] = np.log(df['daily_usdc_volume'])

# Quick sanity check
df['log_usdc_volume'].describe()
```

```python
# run the regression again with the volume as the DV
# Dependent variable
y = df['log_usdc_volume']

# Independent variable
X = df[['eth_vol_7d']]
X = sm.add_constant(X)

# OLS with HC1 robust standard errors
model_vol = sm.OLS(y, X).fit(cov_type='HC1')

print(model_vol.summary())
```

Volume Regression Results

I ran an OLS regression of the logarithm of daily USDC transaction volume on same-day ETH volatility, again using heteroskedasticity-robust (HC1) standard errors. The estimated coefficient on ETH volatility is positive and statistically significant at the 1% level. Interpreted as a semi-elasticity, a 0.01 increase in daily ETH volatility is associated with approximately an 11–12% increase in daily USDC transaction volume.

This regression exhibits low explanatory power (R² = 0.021) as well, which is expected given the highly skewed nature of transaction volumes and the parsimonious design of the model. Taken together, the results suggest that periods of heightened market volatility are associated with increases along both the transaction counts and the transaction volumes of stablecoin usage.

```python
# Create a 1-day lagged ETH volatility
df['eth_vol_7d_lag1'] = df['eth_vol_7d'].shift(1)

# Drop rows where the lagged volatility is undefined
df_lag = df.dropna(subset=['eth_vol_7d_lag1']).copy()

df_lag[['eth_vol_7d', 'eth_vol_7d_lag1']].head()
```

```python
# Dependent variable
y = df_lag['daily_usdc_tx_count']

# Independent variables
X = df_lag[['eth_vol_7d', 'eth_vol_7d_lag1']]
X = sm.add_constant(X)

model_tx_lag = sm.OLS(y, X).fit(cov_type='HC1')

print(model_tx_lag.summary())
```

Robustness: Lagged Volatility (Transaction Counts)

As a robustness check, I re-estimate the transaction-count regression by including a one-day lag of ETH volatility. The coefficient on same-day volatility remains positive and statistically significant at the 5% level, while the coefficient on lagged volatility is positive but not statistically significant. Compared to the baseline model, the estimated effect of same-day volatility is smaller when controlling for yesterday's volatility, reflecting overlap between current and past volatility. Overall, the results indicate that USDC transaction activity is more closely related to current market volatility than to volatility from previous days.

```python
#lagged-volume regression

# Dependent variable: log USDC volume
y = df_lag['log_usdc_volume']

# Independent variables: same-day and lagged volatility
X = df_lag[['eth_vol_7d', 'eth_vol_7d_lag1']]
X = sm.add_constant(X)

# OLS with heteroskedasticity-robust (HC1) standard errors
model_vol_lag = sm.OLS(y, X).fit(cov_type='HC1')

print(model_vol_lag.summary())
```

Robustness: Lagged Volatility (Transaction Volume)

As a robustness check, I re-estimate the volume regression including both same-day and one-day lagged ETH volatility. When both measures are included together, neither coefficient is statistically significant. This suggests that while higher market volatility is associated with increased USDC transaction volumes in the baseline specification, the timing of this relationship is less clearly concentrated on a single day. Overall, the results indicate that stablecoin transaction volumes respond to broader periods of market volatility rather than to short-run day-to-day changes.

```python
#Figure: ETH Volatility and USDC Activity Over Time
# Normalize transaction count for visualization
df['usdc_tx_count_norm'] = (
    df['daily_usdc_tx_count'] - df['daily_usdc_tx_count'].mean()
) / df['daily_usdc_tx_count'].std()

# Normalize ETH volatility for visualization
df['eth_vol_7d_norm'] = (
    df['eth_vol_7d'] - df['eth_vol_7d'].mean()
) / df['eth_vol_7d'].std()

#plot using normalized vars
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

plt.plot(
    df['day'],
    df['eth_vol_7d_norm'],
    label='ETH 7-day Volatility (normalized)',
    linewidth=2
)

plt.plot(
    df['day'],
    df['usdc_tx_count_norm'],
    label='USDC Tx Count (normalized)',
    alpha=0.7
)

plt.xlabel('Date')
plt.ylabel('Standardized Value')
plt.title('ETH Volatility and USDC Transaction Activity Over Time')
plt.legend()

plt.tight_layout()
plt.show()
```

Figure 1. Normalized ETH 7-day realized volatility and normalized daily USDC transaction counts over time. Both series are standardized to facilitate comparison of their time-series dynamics. Periods of elevated market volatility tend to coincide with increased stablecoin transaction activity, consistent with the regression results.

```python
# Day of the week fixed effects

# Day of week: Monday=0, Sunday=6
df['dow'] = df['day'].dt.dayofweek

# dummy variable
dow_dummies = pd.get_dummies(df['dow'], prefix='dow', drop_first=True).astype(float)

#drop missing rows
df_fe = df.dropna(subset=['eth_vol_7d']).copy()

# Dependent variable
y = df_fe['daily_usdc_tx_count']

# Independent variables
X = pd.concat(
    [df_fe[['eth_vol_7d']], dow_dummies.loc[df_fe.index]],
    axis=1
)

X = sm.add_constant(X)

model_tx_dow = sm.OLS(y, X).fit(cov_type='HC1')
print(model_tx_dow.summary())

```

Day-of-week fixed effects

To check whether the results are driven by weekly trading patterns, I re-estimated the model including day-of-week fixed effects. The coefficient on ETH volatility remains positive and statistically significant, with a similar magnitude to the baseline specification. The day-of-week indicators themselves are not statistically significant, suggesting that weekly patterns do not explain the observed relationship. Overall, this supports the robustness of the main result that higher ETH volatility is associated with increased USDC transaction activity.

## III. Limitations
While Phase 2 improves substantially on Phase 1 by using indexed blockchain data, several limitations remain:

    - The analysis is limited to Ethereum mainnet and does not capture stablecoin usage on Layer-2 networks or other blockchains

    - Transfer volume does not distinguish between economic use cases (e.g., payments, exchange settlement, DeFi activity)

    - Market volatility is measured using off-chain price data and may not fully capture on-chain stress conditions

These limitations motivate future extensions using protocol-level data, cross-chain analysis, and richer market indicators.

## IV. Future Research
Future work could incorporate additional on-chain controls such as Ethereum gas fees or overall network activity to better isolate the role of market volatility from network conditions. The analysis could also be extended to other major stablecoins or blockchains to assess whether these results are specific to USDC on Ethereum. Finally, using longer lag structures or event-based approaches around major volatility episodes could help clarify the timing and persistence of stablecoin usage responses.

## V. Conclusion

This notebook examines the relationship between Ethereum price volatility and on-chain USDC transaction activity. Using daily on-chain data from Ethereum and heteroskedasticity-robust regressions, the results show a statistically significant but economically modest relationship between ETH volatility and USDC transaction counts, while effects on transaction volume are weaker and less consistent. Overall, these findings suggest that stablecoin usage may respond to periods of market uncertainty, though volatility alone explains only a small share of observed variation in on-chain activity.
